```python
#define a function to group ages together into ranges.  Under 14 is commonly a definition for dental procedures for children vs adults

def agegroup(x):
    if x == "0-14":
        return "Under 14"
    elif "15"<=x and x<="17":
        return "15-17"
    elif "17"<x and x<="30":
        return "18-30"
    elif "30"<x and x<="40":
        return "31-40"
    elif "40"<x and x<="50":
        return "41-50"
    elif "50"<x and x<="60":
        return "51-60"
    elif "60"<x and x<="64":
        return "61-64"
    elif x > "64":
        return  "Over 64"
    else:
        return "Family"
    
rates['AgeGroup']= rates['Age'].apply(lambda x: agegroup(x))
rates                        
```

```python
ratesshort = rates[['StateCode','PlanId','RatingAreaId','Age', 'AgeGroup','IndividualRate','Couple']]
ratesshort
```
